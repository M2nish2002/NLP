```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df=df.drop(['textID','text'],axis=1)
```

```python
df[df.isna().any(axis=1)]
```

```python
df=df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
import nltk
nltk.download('stopwords')
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
wordlemmatize=WordNetLemmatizer()
```

```python
X_train=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df['selected_text'][i])
    review=review.lower()
    review=review.split()
    review=[wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    X_train.append(review)
```

```python
df1=pd.read_csv('test.csv')
```

```python
df1
```

```python
df1=df1.drop(['textID'],axis=1)
```

```python
df1=df1.dropna()
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
X_test=[]
for i in range(0,len(df1)):
    review=re.sub('[^a-zA-z]',' ',df1['text'][i])
    review=review.lower()
    review=review.split()
    review=[wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    X_test.append(review)
```

```python
sentiment_mapping = {'neutral': 0, 'positive': 1, 'negative': -1}
y_train= df['sentiment'].map(sentiment_mapping).values
```

```python
y_test= df1['sentiment'].map(sentiment_mapping).values
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=2500,ngram_range=(1,3))
```

```python
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
accuracy=accuracy_score(y_test,model.predict(X_test))
```
